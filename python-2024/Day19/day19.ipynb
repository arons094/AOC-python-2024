```python
with open("input19.txt", "r") as f:
    content = f.read().split("\n\n")
    towels = content[0]
    towels = [towel.strip() for towel in towels.split(",")]
    patterns = content[1].split("\n")

```

```python
from functools import cache

@cache
def can_form_design_memo(design, patterns):
    if design == "":
        return True

    for p in patterns:
        if design.startswith(p):
            if can_form_design_memo(design[len(p):], tuple(patterns)):
                return True

    return False

count = 0
for pattern in patterns:
    if can_form_design_memo(pattern, tuple(towels)):
        count += 1
print(f"Part: 1 {count}")
```

```python
#def find_if_pattern_possible(towel, pattern):
def can_form_design(design, patterns):
    n = len(design)
    dp = [False] * (n + 1)
    dp[0] = True  # base case: empty design can be formed trivially

    for i in range(1, n + 1):
        for p in patterns:
            plen = len(p)
            if i - plen >= 0 and dp[i - plen] and design[i - plen:i] == p:
                dp[i] = True
                break  # no need to check further patterns if dp[i] is True

    return dp[n]

count = 0
for pattern in patterns:
    if can_form_design(pattern, towels):
        count += 1
print(f"Part: 1 {count}")
```

```python
from functools import cache

@cache
def count_ways(design, patterns):
    if design == "":
        return 1

    total_ways = 0
    for p in patterns:
        if design.startswith(p):
            total_ways += count_ways(design[len(p):], patterns)

    return total_ways

count = 0
for pattern in patterns:
    count += count_ways(pattern, tuple(towels))
print(f"Part: 2 {count}")
```

```python
def count_ways(design, patterns):
    n = len(design)
    dp = [0] * (n + 1)
    dp[0] = 1  # base case: one way to form empty prefix

    for i in range(1, n + 1):
        for p in patterns:
            plen = len(p)
            if i - plen >= 0 and design[i - plen:i] == p:
                dp[i] += dp[i - plen]

    return dp[n]

count = 0
for pattern in patterns:
    count += count_ways(pattern, towels)
print(f"Part: 2 {count}")
```
