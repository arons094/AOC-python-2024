```python
import heapq

def solve_maze(maze):
    # Directions: N=0, E=1, S=2, W=3
    directions = [(-1, 0), (0, 1), (1, 0), (0, -1)]
    rows = len(maze)
    cols = len(maze[0])

    start = None
    end = None

    # Find start (S) and end (E)
    for r in range(rows):
        for c in range(cols):
            if maze[r][c] == 'S':
                start = (r, c)
            elif maze[r][c] == 'E':
                end = (r, c)

    # Dijkstra forward to get minimal dist to each state
    dist = [[[float('inf')] * 4 for _ in range(cols)] for __ in range(rows)]
    # start facing East (1)
    dist[start[0]][start[1]][1] = 0
    pq = []
    heapq.heappush(pq, (0, start[0], start[1], 1))

    while pq:
        cost, r, c, d = heapq.heappop(pq)
        if cost > dist[r][c][d]:
            continue
        if (r, c) == end:
            # Not stopping here because we need full dist map for all states
            pass

        # Move forward
        fr, fc = r + directions[d][0], c + directions[d][1]
        if 0 <= fr < rows and 0 <= fc < cols and maze[fr][fc] != '#':
            forward_cost = cost + 1
            if forward_cost < dist[fr][fc][d]:
                dist[fr][fc][d] = forward_cost
                heapq.heappush(pq, (forward_cost, fr, fc, d))

        # Turn left
        ld = (d - 1) % 4
        left_cost = cost + 1000
        if left_cost < dist[r][c][ld]:
            dist[r][c][ld] = left_cost
            heapq.heappush(pq, (left_cost, r, c, ld))

        # Turn right
        rd = (d + 1) % 4
        right_cost = cost + 1000
        if right_cost < dist[r][c][rd]:
            dist[r][c][rd] = right_cost
            heapq.heappush(pq, (right_cost, r, c, rd))

    # Find the best cost to reach E
    best_cost = min(dist[end[0]][end[1]])

    # Now we do a backward search to find all states on best paths
    on_best_path = [[False]*cols for _ in range(rows)]
    visited_state = [[[False]*4 for _ in range(cols)] for __ in range(rows)]

    # States with minimal cost at E
    stack = []
    for d in range(4):
        if dist[end[0]][end[1]][d] == best_cost:
            stack.append((end[0], end[1], d))
            visited_state[end[0]][end[1]][d] = True

    # To check predecessors, consider how we can arrive at (r, c, d):
    # 1) By moving forward from (r - dr, c - dc, d)
    #    cost(r - dr, c - dc, d) + 1 = cost(r, c, d)
    # 2) By turning left or right at (r, c, d0):
    #    cost(r, c, d0) + 1000 = cost(r, c, d)
    #    where d0 differs from d by ±1 mod 4

    while stack:
        r, c, d = stack.pop()
        on_best_path[r][c] = True

        current_cost = dist[r][c][d]

        # Check forward predecessor (someone moved forward into (r,c) facing d)
        # If we came from forward, then previous cell = (r - dr, c - dc)
        pr, pc = r - directions[d][0], c - directions[d][1]
        if 0 <= pr < rows and 0 <= pc < cols and maze[pr][pc] != '#':
            # The cost should be dist[pr][pc][d] + 1 = dist[r][c][d]
            if dist[pr][pc][d] + 1 == current_cost:
                if not visited_state[pr][pc][d]:
                    visited_state[pr][pc][d] = True
                    stack.append((pr, pc, d))

        # Check turn predecessors:
        # If we turned left to get d, then original direction was d+1 mod 4
        # cost(r, c, d+1) + 1000 = dist[r][c][d]
        original_left = (d + 1) % 4
        if dist[r][c][original_left] + 1000 == current_cost:
            if not visited_state[r][c][original_left]:
                visited_state[r][c][original_left] = True
                stack.append((r, c, original_left))

        # If we turned right to get d, then original direction was d-1 mod 4
        original_right = (d - 1) % 4
        if dist[r][c][original_right] + 1000 == current_cost:
            if not visited_state[r][c][original_right]:
                visited_state[r][c][original_right] = True
                stack.append((r, c, original_right))

    # Count how many non-wall tiles are on best paths
    count = 0
    for r in range(rows):
        for c in range(cols):
            if maze[r][c] != '#' and on_best_path[r][c]:
                count += 1
    return count

with open("input16.txt") as f:
    maze = [line.strip() for line in f]

print(solve_maze(maze))
```

```python
import heapq
import matplotlib.pyplot as plt
import numpy as np
import imageio
import os

def run_dijkstra(maze):
    rows = len(maze)
    cols = len(maze[0])
    directions = [(-1,0),(0,1),(1,0),(0,-1)]
    
    start = None
    end = None
    for r in range(rows):
        for c in range(cols):
            if maze[r][c] == 'S':
                start = (r,c)
            elif maze[r][c] == 'E':
                end = (r,c)
    
    dist = [[[float('inf')] * 4 for _ in range(cols)] for __ in range(rows)]
    # Start facing East (1)
    dist[start[0]][start[1]][1] = 0
    pq = []
    heapq.heappush(pq, (0, start[0], start[1], 1))
    
    while pq:
        cost, rr, cc, d = heapq.heappop(pq)
        if cost > dist[rr][cc][d]:
            continue
        
        # Move forward
        fr, fc = rr + directions[d][0], cc + directions[d][1]
        if 0 <= fr < rows and 0 <= fc < cols and maze[fr][fc] != '#':
            fwd_cost = cost + 1
            if fwd_cost < dist[fr][fc][d]:
                dist[fr][fc][d] = fwd_cost
                heapq.heappush(pq, (fwd_cost, fr, fc, d))
        
        # Turn left
        ld = (d - 1) % 4
        lcost = cost + 1000
        if lcost < dist[rr][cc][ld]:
            dist[rr][cc][ld] = lcost
            heapq.heappush(pq, (lcost, rr, cc, ld))
        
        # Turn right
        rd = (d + 1) % 4
        rcost = cost + 1000
        if rcost < dist[rr][cc][rd]:
            dist[rr][cc][rd] = rcost
            heapq.heappush(pq, (rcost, rr, cc, rd))
    
    return dist, start, end

def find_best_path_tiles(maze, dist, start, end, output_gif='best_paths_discovery.gif', frame_skip=5):
    rows = len(maze)
    cols = len(maze[0])
    directions = [(-1,0),(0,1),(1,0),(0,-1)]

    # Find best cost at E
    best_cost = min(dist[end[0]][end[1]])

    # We'll do a backward search to find all best path tiles
    on_best_path = [[False]*cols for _ in range(rows)]
    visited_state = [[[False]*4 for _ in range(cols)] for __ in range(rows)]
    
    states = []
    # Initialize with end states that have best_cost
    for d in range(4):
        if dist[end[0]][end[1]][d] == best_cost:
            states.append((end[0], end[1], d))
            visited_state[end[0]][end[1]][d] = True

    # Prepare for visualization
    frames_dir = 'frames_best_paths'
    if not os.path.exists(frames_dir):
        os.makedirs(frames_dir)
    frames = []
    frame_count = 0
    step_counter = 0

    def draw_frame():
        nonlocal frame_count
        img = np.zeros((rows, cols, 3), dtype=np.uint8)
        for rr in range(rows):
            for cc in range(cols):
                ch = maze[rr][cc]
                if ch == '#':
                    # Walls: black
                    img[rr, cc] = [0, 0, 0]
                elif (rr, cc) == start:
                    # Start: green
                    if on_best_path[rr][cc]:
                        img[rr, cc] = [0,255,0]
                    else:
                        img[rr, cc] = [0,255,0] # start always green
                elif (rr, cc) == end:
                    # End: gold if not discovered yet, else green if part of a best path
                    if on_best_path[rr][cc]:
                        img[rr, cc] = [0,255,0]
                    else:
                        img[rr, cc] = [255,215,0]
                else:
                    if on_best_path[rr][cc]:
                        # Best path tile: green
                        img[rr, cc] = [0,255,0]
                    else:
                        # Unvisited space: white
                        img[rr, cc] = [255,255,255]

        fig, ax = plt.subplots(figsize=(5,5), dpi=100)
        ax.imshow(img)
        ax.set_xticks([])
        ax.set_yticks([])
        for spine in ax.spines.values():
            spine.set_visible(False)
        plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
        
        frame_path = os.path.join(frames_dir, f"frame_{frame_count:05d}.png")
        plt.savefig(frame_path, pad_inches=0)
        plt.close(fig)
        frames.append(frame_path)
        frame_count += 1

    # Initial frame
    draw_frame()

    # Backward search to identify all best path tiles
    # We'll process states in a stack-like manner
    while states:
        new_states = []
        # Mark all current states
        for (r, c, d) in states:
            on_best_path[r][c] = True
            current_cost = dist[r][c][d]

            # Forward predecessor
            pr, pc = r - directions[d][0], c - directions[d][1]
            if 0 <= pr < rows and 0 <= pc < cols and maze[pr][pc] != '#':
                if dist[pr][pc][d] + 1 == current_cost:
                    if not visited_state[pr][pc][d]:
                        visited_state[pr][pc][d] = True
                        new_states.append((pr, pc, d))

            # Turn predecessors
            # If turned left to get d, original direction is (d+1)%4
            original_left = (d + 1) % 4
            if dist[r][c][original_left] + 1000 == current_cost:
                if not visited_state[r][c][original_left]:
                    visited_state[r][c][original_left] = True
                    new_states.append((r, c, original_left))

            # If turned right to get d, original direction is (d-1)%4
            original_right = (d - 1) % 4
            if dist[r][c][original_right] + 1000 == current_cost:
                if not visited_state[r][c][original_right]:
                    visited_state[r][c][original_right] = True
                    new_states.append((r, c, original_right))

        states = new_states
        step_counter += 1
        if step_counter % frame_skip == 0:
            draw_frame()

    # Final frame with all best path tiles
    draw_frame()

    # Create GIF
    images = [imageio.imread(f) for f in frames]
    imageio.mimsave(output_gif, images, duration=0.01)
    print(f"GIF saved as {output_gif}")


# Example usage with a small maze:
maze3 = [
"#########",
"#S......#",
"#...##..#",
"#...E...#",
"#.......#",
"#########"
]

dist, start, end = run_dijkstra(maze3)
find_best_path_tiles(maze3, dist, start, end, output_gif='best_paths_discovery2.gif', frame_skip=1)
```

```python
import networkx as nx
import heapq

def solve_with_networkx(maze):
    rows = len(maze)
    cols = len(maze[0])
    
    # Directions: N=0, E=1, S=2, W=3
    directions = [(-1,0),(0,1),(1,0),(0,-1)]
    
    # Find start (S) and end (E)
    start = None
    end = None
    for r in range(rows):
        for c in range(cols):
            if maze[r][c] == 'S':
                start = (r,c)
            elif maze[r][c] == 'E':
                end = (r,c)
    
    # Construct the graph
    G = nx.DiGraph()

    # Add nodes and edges
    for r in range(rows):
        for c in range(cols):
            if maze[r][c] == '#':
                # No states in walls
                continue
            
            # For each direction, this tile can have 4 states
            for d in range(4):
                # Add the node for (r,c,d)
                G.add_node((r,c,d))
                
                # Turning left and right
                left_dir = (d - 1) % 4
                right_dir = (d + 1) % 4
                G.add_edge((r,c,d), (r,c,left_dir), weight=1000)
                G.add_edge((r,c,d), (r,c,right_dir), weight=1000)
                
                # Moving forward
                dr, dc = directions[d]
                fr, fc = r + dr, c + dc
                if 0 <= fr < rows and 0 <= fc < cols and maze[fr][fc] != '#':
                    # Forward edge
                    G.add_edge((r,c,d), (fr,fc,d), weight=1)

    # We start at S facing East (d=1)
    start_state = (start[0], start[1], 1)
    
    # Use Dijkstra's algorithm from networkx
    dist, _ = nx.single_source_dijkstra(G, start_state)
    
    # Find the best cost at end tile for any direction
    best_cost = float('inf')
    for d in range(4):
        if (end[0], end[1], d) in dist:
            if dist[(end[0], end[1], d)] < best_cost:
                best_cost = dist[(end[0], end[1], d)]
    
    return best_cost if best_cost != float('inf') else None

# Example usage:
maze4 = [
"###############",
"#.......#....E#",
"#.#.###.#.###.#",
"#.....#.#...#.#",
"#.###.#####.#.#",
"#.#.#.......#.#",
"#.#.#####.###.#",
"#...........#.#",
"###.#.#####.#.#",
"#...#.....#.#.#",
"#.#.#.###.#.#.#",
"#.....#...#.#.#",
"#.###.#.#.#.#.#",
"#S..#.....#...#",
"###############"
]

result = solve_with_networkx(maze)
print("Minimal cost:", result)
```
