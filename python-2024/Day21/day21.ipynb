```python
#!/usr/bin/env python3

import sys
import re
from itertools import permutations

# Read lines from stdin, filter for lines matching /\d{3}A/
with open('input21.txt') as f:
    codes = [line.strip() for line in f]


# Replace Ruby's global $positions with a Python dictionary
# We use tuples (x, y) or (row, col); must be consistent throughout.
POSITIONS = {
    '7': (0, 0),
    '8': (0, 1),
    '9': (0, 2),
    '4': (1, 0),
    '5': (1, 1),
    '6': (1, 2),
    '1': (2, 0),
    '2': (2, 1),
    '3': (2, 2),
    '0': (3, 1),
    'A': (3, 2),
    '^': (0, 1),
    'a': (0, 2),
    '<': (1, 0),
    'v': (1, 1),
    '>': (1, 2),
}

# Directions dictionary: how each symbol changes (x, y)
DIRECTIONS = {
    '^': (-1, 0),
    'v': (1, 0),
    '<': (0, -1),
    '>': (0, 1),
}

# Python equivalent of Rubyâ€™s se_to_moveset(start, fin, avoid=Vector[0, 0])
def se_to_moveset(start, fin, avoid=(0, 0)):
    """
    Build all permutations of a path made of '^', 'v', '<', '>', such that:
    1) The net effect moves from 'start' to 'fin'.
    2) Avoid stepping on 'avoid'.
    3) Each valid permutation is appended with 'a'.
    If no valid permutations, return ['a'].
    """
    dx = fin[0] - start[0]
    dy = fin[1] - start[1]

    # Construct a string like '^^^' or 'vv' or '<<' etc.
    path_string = ''
    if dx < 0:
        path_string += '^' * abs(dx)
    else:
        path_string += 'v' * dx
    if dy < 0:
        path_string += '<' * abs(dy)
    else:
        path_string += '>' * dy

    # Permute all possible orders of those arrow characters
    chars = list(path_string)
    unique_perms = set(permutations(chars, len(chars)))

    rv = []
    for perm in unique_perms:
        # Build the sequence of intermediate positions
        pos_list = [start]
        valid = True
        for symbol in perm:
            dxy = DIRECTIONS[symbol]
            next_pos = (pos_list[-1][0] + dxy[0],
                        pos_list[-1][1] + dxy[1])
            pos_list.append(next_pos)
            if next_pos == avoid:
                valid = False
                break
        if valid:
            # Join the symbols back + 'a'
            rv.append("".join(perm) + 'a')

    # If no valid permutations, fallback to ['a']
    if not rv:
        rv = ['a']

    return rv

# Mimic Rubyâ€™s $ml_memos (global memo for min_length)
ML_MEMOS = {}

# Python version of min_length(str, lim=2, depth=0)
def min_length(s, lim=2, depth=0):
    """
    Recursively compute a "length" measure for the given string 's',
    with possible "movesets" and memoization.
    """
    memo_key = (s, depth, lim)
    if memo_key in ML_MEMOS:
        return ML_MEMOS[memo_key]

    # Ruby logic:
    # avoid = depth == 0 ? Vector[3, 0] : Vector[0, 0]
    # cur   = depth == 0 ? $positions['A'] : $positions['a']
    if depth == 0:
        avoid = (3, 0)
        cur = POSITIONS['A']
    else:
        avoid = (0, 0)
        cur = POSITIONS['a']

    length = 0
    for char in s:
        next_cur = POSITIONS[char]
        moveset = se_to_moveset(cur, next_cur, avoid)
        if depth == lim:
            # length += (moveset[0] || 'a').size
            # in Python, moveset[0] or 'a'
            chosen = moveset[0] if moveset else 'a'
            length += len(chosen)
        else:
            # length += moveset.map {min_length(_1, lim, depth + 1)}.min
            # so we recursively call min_length on each possible moveset string
            sub_lengths = []
            for mv in moveset:
                sub_lengths.append(min_length(mv, lim, depth + 1))
            length += min(sub_lengths)
        cur = next_cur

    ML_MEMOS[memo_key] = length
    return length

# Finally, replicate the summation logic
complexity_a = 0
complexity_b = 0

for code in codes:
    length_a = min_length(code)       # lim=2 by default
    length_b = min_length(code, 25)   # lim=25
    numeric = int(code[0:3])         # first 3 digits
    complexity_a += length_a * numeric
    complexity_b += length_b * numeric

print(f"Part 1: {complexity_a}")
print(f"Part 2: {complexity_b}")
```

```python
#!/usr/bin/env python3

import sys
import re
from itertools import permutations
from functools import cache

with open('input21.txt') as f:
    codes = [line.strip() for line in f]

POSITIONS = {
    '7': (0, 0),
    '8': (0, 1),
    '9': (0, 2),
    '4': (1, 0),
    '5': (1, 1),
    '6': (1, 2),
    '1': (2, 0),
    '2': (2, 1),
    '3': (2, 2),
    '0': (3, 1),
    'A': (3, 2),
    '^': (0, 1),
    'a': (0, 2),
    '<': (1, 0),
    'v': (1, 1),
    '>': (1, 2),
}

DIRECTIONS = {
    '^': (-1, 0),
    'v': (1, 0),
    '<': (0, -1),
    '>': (0, 1),
}

def se_to_moveset(start, fin, avoid=(0, 0)):
    dx = fin[0] - start[0]
    dy = fin[1] - start[1]

    # Create a base string of arrow movements:
    path_string = ''
    if dx < 0:
        path_string += '^' * abs(dx)
    else:
        path_string += 'v' * dx

    if dy < 0:
        path_string += '<' * abs(dy)
    else:
        path_string += '>' * dy

    # Permute all possible orders
    chars = list(path_string)
    unique_perms = set(permutations(chars, len(chars)))

    rv = []
    for perm in unique_perms:
        pos_list = [start]
        valid = True
        for symbol in perm:
            dxy = DIRECTIONS[symbol]
            next_pos = (pos_list[-1][0] + dxy[0],
                        pos_list[-1][1] + dxy[1])
            pos_list.append(next_pos)
            if next_pos == avoid:
                valid = False
                break
        if valid:
            rv.append("".join(perm) + 'a')

    if not rv:
        rv = ['a']

    return rv

@cache
def min_length(s, lim=2, depth=0):
    """
    Recursively compute a "length" measure for the given string 's'.
    depth=0 => start on 'A' (avoid=(3,0)).
    depth>0 => start on 'a' (avoid=(0,0)).
    """
    if depth == 0:
        avoid = (3, 0)
        cur = POSITIONS['A']
    else:
        avoid = (0, 0)
        cur = POSITIONS['a']

    length = 0
    for char in s:
        next_cur = POSITIONS[char]
        moveset = se_to_moveset(cur, next_cur, avoid)
        if depth == lim:
            chosen = moveset[0] if moveset else 'a'
            length += len(chosen)
        else:
            sub_lengths = []
            for mv in moveset:
                sub_lengths.append(min_length(mv, lim, depth + 1))
            length += min(sub_lengths)
        cur = next_cur

    return length

complexity_a = 0
complexity_b = 0
for code in codes:
    length_a = min_length(code)       # Default lim=2
    length_b = min_length(code, 25)   # lim=25
    numeric = int(code[0:3])         # first 3 digits
    complexity_a += length_a * numeric
    complexity_b += length_b * numeric

print(f"Part 1: {complexity_a}")
print(f"Part 2: {complexity_b}")
```
