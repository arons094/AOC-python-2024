````python
def find_paths(grid):
    rows, cols = len(grid), len(grid[0])
    paths = []
    
    def is_valid(x, y, current_value):
        return 0 <= x < rows and 0 <= y < cols and grid[x][y] == current_value + 1
    
    def dfs(x, y, current_path):
        # Base case: If we reach the value 9, store the path
        if grid[x][y] == 9:
            paths.append(current_path[:])
            return
        
        # Possible moves: up, down, left, right
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        current_value = grid[x][y]
        
        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if is_valid(nx, ny, current_value):
                dfs(nx, ny, current_path + [(nx, ny)])
    
    # Find all starting points with value 0
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == 0:
                dfs(i, j, [(i, j)])
    
    return paths


# Example usage:
with open("./input10.txt", 'r') as f:
    lines = f.readlines()
    grid = [[int(c) for c in line.strip()] for line in lines]

result = find_paths(grid)
print(len(result))  # 2
````
