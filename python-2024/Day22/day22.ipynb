```python
from collections import defaultdict
from itertools import pairwise, windowed


def f(s):
    s = (s ^ s << 6) & 0xFFFFFF
    s = (s ^ s >> 5) & 0xFFFFFF
    return (s ^ s << 11) & 0xFFFFFF


ans1, ans2 = 0, defaultdict(int)

for s in map(int, open('input22.txt')):
    nums = [s]+[s:=f(s) for _ in range(2000)]
    ans1 += nums[-1]

    diffs = [b%10-a%10 for a,b in pairwise(nums)]

    seen = set()
    for i in range(len(diffs)-3):
        pat = tuple(diffs[i:i+4])
        if pat not in seen:
            ans2[pat] += nums[i+4] % 10
            seen.add(pat)

print(ans1, max(ans2.values()))
```

```python
from collections import defaultdict
from itertools import pairwise

with open('input22.txt') as f:
    secrets = [int(x) for x in f.readlines()]

def secret_gen(secret_0):
    secret_1 = (secret_0 ^ secret_0 << 6) & 0xFFFFFF
    secret_2 = (secret_1 ^ secret_1 >> 5) & 0xFFFFFF
    new_secret = (secret_2 ^ secret_2 << 11) & 0xFFFFFF
    return new_secret

ans = 0
sell_values = defaultdict(int)
for s in secrets:
    nums = [s]
    nums.extend([s:=secret_gen(s) for _ in range(2000)])
    ans += nums[-1]

    diffs = [b%10 - a%10 for a,b in pairwise(nums)]
    seen = set()
    for i in range(len(diffs)-3):
        pattern = tuple(diffs[i:i+4])
        if pattern not in seen:
            sell_values[pattern] += nums[i+4] % 10
            seen.add(pattern)
ans2 = max(sell_values.values())

print(f"Part1: {ans}, Part2: {ans2}")
```
