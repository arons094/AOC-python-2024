```python

with open("input17.txt") as f:
    data = f.read().strip()
lines = data.split("\n")
a = int(lines[0].split()[-1])
b = int(lines[1].split()[-1])
c = int(lines[2].split()[-1])
program = list(map(int, lines[4].split(": ")[1].split(",")))
```

```python
def combo(a, b, c, value):
    if value < 4:
        return value
    if value == 4:
        return a
    if value == 5:
        return b
    if value == 6:
        return c
    return None

def evaluate(a, b, c, ip, program, logging=False):
    opcode = program[ip]
    arg = program[ip+1]
    com = combo(a, b, c, arg)
    a, b, c, = int(a), int(b), int(c)
    if logging:
        print(f"Executing: {opcode} Arg: {arg} Combo: {com}")
    if opcode == 0:
        numerator = a
        denominator = 2**com
        new_a = numerator // denominator
        if logging:
            print(f"Setting A to {new_a}")
        return (None, new_a, b, c, ip+2)
    elif opcode == 1:
        new_b = b^arg
        if logging:
            print(f"Setting B to {new_b}")
        return (None, a, new_b, c, ip+2)
    elif opcode == 2:
        new_b = com % 8
        if logging:
            print(f"Setting B to {new_b}")
        return (None, a, new_b, c, ip+2)
    elif opcode == 3:
        if a == 0:
            if logging:
                print(f"Jumping to {ip+2}")
            return (None, a, b, c, ip+2)
        else:
            if logging:
                print(f"Jumping to {arg}")
            return (None, a, b, c, arg)
    elif opcode == 4:
        new_b = b^c
        if logging:
            print(f"Setting B to {new_b}")
        return (None, a, new_b, c, ip+2)
    elif opcode == 5:
        if logging:
            print(f"Returning {com % 8}")
        return (com % 8, a, b, c, ip+2)
    elif opcode == 6:
        numerator = a
        denominator = 2**com
        new_b = numerator // denominator
        if logging:
            print(f"Setting B to {new_b}")
        return (None, a, new_b, c, ip+2)
    elif opcode == 7:
        numerator = a
        denominator = 2**com
        new_c = numerator // denominator
        if logging:
            print(f"Setting C to {new_c}")
        return (None, a, b, new_c, ip+2)

    
def run_program(a, b, c, program, logging=False):
    ip = 0
    result = []
    while ip < len(program) - 1:
        (out, a, b, c, ip) = evaluate(a, b, c, ip, program, logging)
        if out is not None:
            result.append(out)
    return result

def remake_self(program, cur, part):
    for candidate in range(8):
        try_A_val = part * 8 + candidate
        if run_program(try_A_val, 0, 0, program) == program[cur:]:
            if cur == 0:
                return try_A_val
            nxt = remake_self(program, cur-1, try_A_val)
            if nxt is not None:
                return nxt
    return None
ans = ",".join([str(x) for x in run_program(a, b, c, program)])
print(ans)
ans2 = remake_self(program, len(program)-1, 0)
print(ans2)
```
